```python
from openpyxl import Workbook, load_workbook
from openpyxl.drawing.image import Image
from openpyxl.utils.dataframe import dataframe_to_rows
j=0
res_football=[]
res_volleyball=[]
res_basketball=[]
comp_fotball=[]
comp_vol=[]
comp_bas=[]
year_fot=[]
year_vol=[]
year_bas=[]
for i in table.Competition:
    df=pd.read_html(table['html'][j])[0]
    if('بسکتبال' in i):
        df.columns=["level", "team","game","win","lose","scores achieved", "scores lost", "score difference","points","Unnamed: 9","Unnamed: 10"]
        res_basketball.append(df)
        for m in range(len(df)):
            comp_bas.append(table.Competition[j])
            year_bas.append(table.Year[j])
        
    elif('والیبال' in i):
        df.columns=["level", "team", "point", "game", "win", "lose", "30", "31", "32", "23", "13", "03", "won set", "lost set", "average set", "won point", "lost point", "average point"]
        res_volleyball.append(df)
        for m in range(len(df)):
            comp_vol.append(table.Competition[j])
            year_vol.append(table.Year[j])
        
    else:
        df.columns=["level", "team", "plays", "win", "eq", "lose", "scores achieved", "scores lost", "score difference", "point", "unnamed"]        
        res_football.append(df)
        for m in range(len(df)):
            comp_fotball.append(table.Competition[j])
            year_fot.append(table.Year[j])
    j+=1
#print(comp_fotball)
football_futsal_df = pd.concat(res_football).reset_index()
football_futsal_df.drop("index", inplace=True, axis=1)
football_futsal_df.drop("unnamed", inplace=True, axis=1)
football_futsal_df.insert(0, "competation", comp_fotball, True) 
football_futsal_df.insert(1, "year", year_fot, True) 
df = pd.DataFrame(data=football_futsal_df)
df = df.fillna(0) 
wb = load_workbook('table_football_footsal_finall.xlsx')
active = wb['Sheet1']
for x in dataframe_to_rows(df):
    active.append(x)    
wb.save('table_football_footsal_finall.xlsx')
wb.close()

football_futsal_df = pd.concat(res_volleyball).reset_index()
football_futsal_df.drop("index", inplace=True, axis=1)
football_futsal_df.insert(0, "competation", comp_vol, True) 
football_futsal_df.insert(1, "year", year_vol, True) 
df = pd.DataFrame(data=football_futsal_df)
df = df.fillna(0) 
wb = load_workbook('table_volleyball_finall.xlsx')
active = wb['Sheet1']
for x in dataframe_to_rows(df):
    active.append(x)    
wb.save('table_volleyball_finall.xlsx')
wb.close()

football_futsal_df = pd.concat(res_basketball).reset_index()
football_futsal_df.drop("index", inplace=True, axis=1)
football_futsal_df.drop("Unnamed: 9", inplace=True, axis=1)
football_futsal_df.drop("Unnamed: 10", inplace=True, axis=1)
football_futsal_df.insert(0, "competation", comp_bas, True) 
football_futsal_df.insert(1, "year", year_bas, True) 
df = pd.DataFrame(data=football_futsal_df)
df = df.fillna(0) 
wb = load_workbook('table_basketball_finall.xlsx')
active = wb['Sheet1']
for x in dataframe_to_rows(df):
    active.append(x)    
wb.save('table_basketball_finall.xlsx')
wb.close()
```
